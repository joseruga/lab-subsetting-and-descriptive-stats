# Challenge 1

#### In this challenge we will use the `Temp_States.csv` file. 

#### First import it into a data frame called `temp`.

```python
# la funcion pd.read_csv(<'ruta'>), abre cualquier liga, se le debe pasar en una str la ruta del archivo

temp = pd.read_csv('Temp_States.csv')
```

#### Print `temp`

```python
# Revisamos a grandes rasgos los datos

print(temp)
```

#### Explore the data types of the Temp dataframe. What type of data do we have? Comment your result.

```python
# <DF>.dtypes es una propiedad que nos permite revisar el tipo d datos de un DF, el nombre debe ir antes

temp.dtypes
```

#### Select the rows where state is New York

```python
# Recordar primero poner el df, luego poner la columna (referencianda desde el df) y finalmente la prueba logica

# <DF>[<DF>[<'Columna a filtrar'>] == 'Comparacion']

temp[temp['State']=='New York']
```

#### What is the average of the temperature of cities in New York?

```python
# Primero filtramos solo NewYork, como en la pregunta anterior
# Despues de haber seleccionado el filtro con la prueba logica agregamos solo la columna que queremos mostrar
# Finalmente usamos el metodo .mean() de pandas que nos regresa el promedio de esa serie o DF
# <DF>[<DF>[<'Columna a filtrar'>] == 'Comparacion']['Columna a mostrar'].mean()

temp[temp['State']=='New York']['Temperature'].mean()
```

#### We want to know cities and states with Temperature above 15 degress Celcius

```python
# Filtramos ahora con una prueba logica numerica
# Ya que no pide mostrar columnas especificas no es necesario agregar las columnas a mostrar
# <DF>[<DF>[<'Columna a filtrar'>] > <valor numerico>]

temp[temp['Temperature']>15]
```

#### Now, return only the cities that have a temperature above 15 degress Celcius

```python
# El filtro es igual a la pregunta anterior
# Ya que nos piden mostrar solo las ciudades, especificamos las columnas a mostrar
# <DF>[<DF>[<'Columna a filtrar'>] > <valor numerico>][<'Columna a mostrar'>]

temp[temp['Temperature']>15]['City']
```

#### We want to know which cities have a temperature above 15 degrees Celcius and below 20 degrees Celcius

*Hint: First write the condition then select the rows.*

```python
# Para filtrar por multiples condiciones debemos usar & o | y cada condicion debe estar entre parentesis
# <DF>[(condicion_1) & (condicion_2)][<'Columnas a mostrar'>]

temp[(temp['Temperature']>15) & (temp['Temperature']<20)]['City']
```

#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
# Utilizamos el metodo .groupby('columna a agrupar'), en este caso state
# Utilizamos el metoo .agg({'columna':'agregacion'}) para crear columnas calculadas por la agregacion del groupby
# Si queremos que diferentes agregaciones se hagan para la misma columna en los valores del dict pasamos una lista con éstas

temp.groupby('State').agg({'Temperature':['mean','std']})
```

#### Print the first 3 rows of your dataframe

##### Tip : There are 2 ways to do it. Do it both ways

```python
# Una manera es utilizar el metodo DF.head() y pasar como parametro tres, para ver tres filas

employee.head(3)
```

```python
# Otra manera es hacerlo por indices, similar a una lista

employee[:3]
```

```python
# Para este DF en especifico que los indexes son numeros de 0 a n podemos usar el metodo DF.loc()
# DF.loc() funciona con el nombre de los indexes y columnas, en este caso los nombres son numeros
# Igual que con DF.iloc() este hay diferentes sintaxis que daran el mismo resultado:
# ¡OJO! ya que aqui es por el nombre y no el index, debemos poner hasta el 2 ya que es incluyente 

#Sintaxis uno
employee.loc[:2]

#Sintaxis dos
employee.loc[:2,:]
```

#### Find the employees who's title is associate and the salary above 55?

```python
# Utilizamos dos condiciones logicas, recordar poner cada condicion entre () y usar & en lugar de 'and'

# DF[(<condicion 1>) & (<condicion 2>)]

employee [ (employee ['Title'] == 'associate') & (employee ['Salary'] > 55)]
```

#### Group the employees based on their number of years of employment. What are the average salaries in each group?

```python
# Hay un par de formas de hacer esto aqui
# Utilizamos groupby y aplicamos el metodo DF.mean(), despues especificamos la columna sobre la cual calcular
# DF.groupby(<'Columna a agrupar'>).mean()[<'Columna a calcular'>]

employee.groupby('Years').mean()['Salary']

# Otra manera de hacerlo es usar el metodo DF.agg() y especificar la columna y metodo de agregacion

employee.groupby('Years').agg({'Salary':'mean'})
```

####  What is the average Salary per title?

```python
# Similar al problema anterior, hay dos metodos de hacerlo

employee.groupby('Title').agg({'Salary':'mean'})

employee.groupby('Title').mean()['Salary']
```
